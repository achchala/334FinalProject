# Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(pd.read_excel('334data2.xlsx')) 
df.head()
```

```python
print('There are ', len(df), ' rows of data.')
```

# Time Series Forecasting

```python
# Assuming the monthly data starts from the 10th column onwards based on the data sample structure
monthly_data = df.iloc[:, 10:22]

# Prepare the time variable (month number) as the predictor
# Considering months from January (1) to December (12) for 2019
months = np.arange(1, 13).reshape(-1, 1)

# Since the file appears to contain multiple rows/products, let's focus on the first product for this demonstration
# Extracting the demand data for the first product
demand_data = monthly_data.iloc[0].values.reshape(-1, 1)

# Initialize and fit the Linear Regression model
model = LinearRegression()
model.fit(months, demand_data)

# Coefficients of the model
slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```
